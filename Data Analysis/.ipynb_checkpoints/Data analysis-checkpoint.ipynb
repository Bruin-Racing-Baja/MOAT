```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_log = np.loadtxt('test_data.txt', dtype=float, delimiter=', ', skiprows=4, ndmin=2)
print(data_log)
```

```python
# status = data_log[0]
# rpm = data_log[1]
# act_vel = data_log[2]
# enc_pos = data_log[3]
# full_in = data_log[4]
# full_out = data_log[5]
# control_start = data_log[6]
# control_stop = data_log[7]
# odrive_voltage = data_log[8]
# odrive_current = data_log[9]
# cooling_temp = data_log[10]

print("everything is fine")
```

```python
for i in range(len(data_log)): # number of rows
    status[i] = data_log[i][0]
    rpm[i] = data_log[i][1]
    act_vel[i] = data_log[i][2]
    enc_pos = data_log[i][3]
    full_in = data_log[i][4]
    full_out = data_log[i][5]
    control_start = data_log[i][6]
    control_stop = data_log[i][7]
    odrive_voltage = data_log[i][8]
    odrive_current = data_log[i][9]
    cooling_temp = data_log[i][10]
```

```python
# #Array of graphs
# row_num = 2
# col_num = 3

# plt.subplot(row_num, col_num, 1) 
plt.plot(control_start, enc_pos)
print(enc_pos)
plt.title("Encoder position")
plt.xlabel('Control start time (ms)')
plt.ylabel('Encoder pos (ticks)')
```

```python

#Plot on top of each other
#plt.subplot(row_num, col_num, 2) 
plt.plot(control_start, odrive_voltage, color='r', label='voltage')
plt.plot(control_start, odrive_current, color='g', label='current')
plt.title("ODrive voltage and current")
plt.xlabel('Control start time (ms)')
plt.legend()
```

```python

#plt.subplot(row_num, col_num, 4) 
plt.plot(control_start, rpm)
plt.title("Motor RPM")
plt.xlabel('Control start time (ms)')
plt.ylabel('Motor RPM (rot/min)')
```

```python

# plt.subplot(row_num, col_num, 5) 
plt.plot(control_start, act_vel)
plt.title("Actuator velocity")
plt.xlabel('Control start time (ms)')
plt.ylabel('Actuator velocity (m/s)')
```

```python

# plt.subplot(row_num, col_num, 6) 
plt.plot(control_start, cooling_temp)
plt.title("Cooling temperature")
plt.xlabel('Control start time (ms)')
plt.ylabel('Cooling temperature (deg C)')

```
